```python
class KMean():
    
##########################Intilising######################   
    def __init__(self,X, **kwargs):
       
        self.XO=np.zeros(len(X))
        self.X=np.column_stack([X,self.XO])
        self.no_K=2
        self.iterations=5
        self.plot=True
        self.Z=-999

        
        ##########Override Default parameters######
        
        self.args = ['no_K', 'iterations','plot']
        
        for a in self.args:
            if a in kwargs:
                exec('self.'+a+' = kwargs[a]')
        
        
        self.kmeans=KMean.__kcalc(self.no_K)
        
     
    
#####################Calculate########################  

    def Run(self):
        if  self.plot==True:
            KMean.__plot(self.kmeans,self.X,'Run 0 : Initial positions of centroids')
        for i in range( self.iterations):     
            self.X=KMean.__allocate_k(self.kmeans,self.X)    
            self.kmeans=KMean.__calculate_k(self.kmeans,self.X)
            if  self.plot==True:
                KMean.__plot(self.kmeans,self.X,'Run '+str(i+1)+' : Realigned centroids')
            
            
#####################GET##########################################


    def Get_Output(self):
        return self.X
    
    def Get_Kmeans(self):
        print(self.kmeans)
        
    def Get_Parameters(self):  
        for a in self.args:
            exec('self.Z; self.Z=self.'+a)
            print(a+': '+str(self.Z))
    
##########################Internal Functions######################    
    
    def __allocate_k(kmeans,X):
        for x in X:
            cost=np.zeros(len(kmeans))
            for i,k in enumerate(kmeans):
                c=(np.linalg.norm(x[:2]-k))**2
                cost[i]=c
            x[2]=cost.argmin()
        return X


    def __calculate_k(kmeans,X):
        for i,k in enumerate(kmeans):

            for n in range(len(k)):
                last=k[n]
                k[n]=X[:,n][X[:,-1]==i].sum()/len(X[:,-1][X[:,-1]==i])
                if np.isnan(k[n]):
                    k[n]=last

            kmeans[i]=k  

        return kmeans
  
    
        
    def __plot(kmeans,X,Tit):
        import matplotlib.cm as cm
        import matplotlib.pyplot as plt
        N=range(len(kmeans))
        col = cm.rainbow(np.linspace(0, 1, len(kmeans)))
        for n,c in zip(N,col):
            CX=X[X[:,2]==n]    
            plt.scatter(CX[:,0],CX[:,1],color=c)    

        plt.scatter(kmeans[:,0],kmeans[:,1] ,color='black', marker="x")
        plt.title(Tit)
        plt.show()    
        
        
        
        
    def __kcalc(no_K):
        for i in range(no_K):
            if i==0:
                kmeans=np.array([[np.random.normal(),np.random.normal()]])
            else:
                kmeans=np.append(kmeans,[[np.random.normal(),np.random.normal()]], axis=0)
        return kmeans

    
```
